```python
param = {
    "key": api_key,
    "part": "snippet",
    "q": "Nollywood",
    "type": "video",
    "maxResults": 45
}

request = requests.get(search_url, params=param).json()

data = []

for item in request['items']:
    video_params = {
        "key": api_key,
        "part": "statistics",
        "id": item['id']['videoId']
    }
    video_request = requests.get(videos_url, params=video_params).json()

    publishdate = item['snippet']['publishedAt']
    title = item['snippet']['title']
    time = item['snippet']['publishTime']
    channel_title = item['snippet']['channelTitle']
    url = "www.youtube.com/watch?v=" + item['id']['videoId']
    view_count = video_request['items'][0]['statistics']['viewCount']
    like_count = video_request['items'][0]['statistics']['likeCount']
    
    # Append the extracted values to the data list
    data.append({
        'publishdate': publishdate,
        'title': title,
        'time': time,
        'channel_title': channel_title,
        'url': url,
        'view_count': view_count,
        'like_count': like_count
    })

# Create a DataFrame from the data list
youtube_datas = pd.DataFrame(data)

# Print the resulting DataFrame
print(youtube_datas)
```

```python
 youtube_datas.to_csv("Nollywood_movi2.csv")
```

```python


search_params = {
    "key": api_key,
    "part": "snippet",
    "q": "Nollywood",
    "type": "video",
    "maxResults": 50  # Maximum allowed by the API
}

# Send the first search request
data = []
next_page_token = None
while True:
    if next_page_token:
        search_params['pageToken'] = next_page_token
    request = requests.get(search_url, params=search_params).json()

    # Extract video data from the search results
    for item in request['items']:
        video_params = {
            "key": api_key,
            "part": "statistics",
            "id": item['id']['videoId']
        }
        video_request = requests.get(videos_url, params=video_params).json()

        publishdate = item['snippet']['publishedAt']
        title = item['snippet']['title']
        time = item['snippet']['publishTime']
        channel_title = item['snippet']['channelTitle']
        url = "www.youtube.com/watch?v=" + item['id']['videoId']
        view_count = video_request['items'][0]['statistics']['viewCount']
        like_count = video_request['items'][0]['statistics'].get('likeCount', 0)

        # Append the extracted values to the data list
        data.append({
            'publishdate': publishdate,
            'title': title,
            'time': time,
            'channel_title': channel_title,
            'url': url,
            'view_count': view_count,
            'like_count': like_count
        })

    # Check if there are more search results
    if 'nextPageToken' in request:
        next_page_token = request['nextPageToken']
    else:
        break

# Create a DataFrame from the data list
youtube_datas = pd.DataFrame(data)

# Print the resulting DataFrame
print(youtube_datas)
```

```python
 youtube_datas.to_csv("Nollywood_full_movie.csv")
```
